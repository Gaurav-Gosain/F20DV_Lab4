```python
import pandas as pd
import json

df = pd.read_csv('Hotel_Reviews.csv')

with open('europe.geojson') as f:
    j = json.load(f)

data = pd.read_json(json.dumps([i['properties'] for i in j['features']]))

data
```

```python
# Loop through each row in the dataframe
for i in range(len(data)):
    
    temp = df[df['Hotel_Address'].str.contains(data.iloc[i]
          ['NAME'].replace("(", "").replace(")", ""))]

    if temp.shape[0]:
        print(data.iloc[i]['NAME'])
        temp.to_csv(f'data/{data.iloc[i].ISO3}.csv', index=False)
```
